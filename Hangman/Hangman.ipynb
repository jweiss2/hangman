```python
import random as rand
```

```python
file = open("C:\\users\jonat\Documents\Python Scripts\Hangman\wordlist.txt")
words = file.read()
file.close()

words = words.split()
words = [word for word in words if len(word) > 3];
```

```python
life = iter(['\
 ___\n\
/   \\\n\
\   /\n\
 \u203E\u203E\u203E\n','\
 ___\n\
/   \\\n\
\   /\n\
 \u203E\u203E\u203E\n\
  |\n','\
 ___\n\
/   \\\n\
\   /\n\
 \u203E\u203E\u203E\n\
  |__\n','\
 ___\n\
/   \\\n\
\   /\n\
 \u203E\u203E\u203E\n\
__|__\n','\
 ___\n\
/   \\\n\
\   /\n\
 \u203E\u203E\u203E\n\
__|__\n\
  |\n','\
 ___\n\
/   \\\n\
\   /\n\
 \u203E\u203E\u203E\n\
__|__\n\
  |\n\
 /\n\
/\n','\
 ___\n\
/   \\\n\
\   /\n\
 \u203E\u203E\u203E\n\
__|__\n\
  |\n\
 / \ \n\
/   \ \n'
        
            ])
```

```python
restart = True

while restart == True:
    restart = False
    # Select random word
    ind = rand.randint(0,len(words)-1)
    word = words[ind]

    display = ['_ '] * len(word)
    print(''.join(display))

    #keep track of guesses
    record = []

    while 1:

        guess = input('Guess a letter: ').lower();

        # check if entered letter
        if guess.isalpha() == False:
            print("Please enter a letter.\n")
            continue
        elif guess in record:
            print("Already guessed this letter.\n")
            continue

        # add guess to record
        record.append(guess)

        # loop through letters
        count = 0 # for index
        corr  = False # did you make a good guess?
        for letter in word:
            if guess == letter:

                display[count] = guess + ' '
                corr = True
            count += 1

        # Wrong guess
        if corr == False:
            # print hangman
            try: 
                print(next(life))
            except StopIteration: 
                print("Game Over.")
                break
            print('Bad guess, try again')

        print(''.join(display))

        # end while loop if entire word is guessed
        if ('_ ' in display) == False:
            print("Good job my guy!")

            #restart?
            again = input("Press 'y' to play again.").lower()
            if again == 'y':
                restart = True

            break
```
